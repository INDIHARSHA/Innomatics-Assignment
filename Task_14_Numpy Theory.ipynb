### Create 2-D and 3-D arrays. And check their shape and dimensions using (shape and ndim methods).

```python
import numpy as np
arr1=np.array([[1,5,4,6],[4,5,5,5]])
print(arr1.shape,arr1.ndim)
arr2=np.array([[[1,5,4,6],
                [4,5,5,5],
                [5,84,15,25]]])
print(arr2.shape,arr2.ndim)
```

## Creation of numpy with different methods:

### Create a 1D NumPy array with elements from 1 to 10 using arange().

```python
import numpy as np
ele=np.arange(1,11)
print(ele,ele.ndim)
```

### Create a 2D array of size 3x3 filled with zeros using zeros().

```python
elezero=np.zeros((3,3))
print(elezero,elezero.shape)
```

### Create an array [5, 10, 15, 20, 25]. Find its size, item size, and total bytes used by the array.

```python
anarray=np.array([5, 10, 15, 20, 25])
print(f"array size:{anarray.size} '\n'{anarray.itemsize},'\n',{anarray.tobytes}")
```

### Use linspace() to generate an array of 7 equally spaced numbers between 0 and 14.

### Task

* If  is odd, print Weird
* If  is even and in the inclusive range of  2-5, print Not Weird
* If  is even and in the inclusive range of  6-20 , print Weird
* If  is even and greater than , print Not Weird

```python
n = int(input())
if (n%2==0 and n in range(2,6)) or (n%2==0 and n>20):
    print("Not Weird")
elif (n%2!=0)or(n%2==0 and n in range(6,21)):
    print("Weird")
else:
    print("Enter a valid number")
```

### 6)Task

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function.

```python
def is_leap(year):
    leap = False
    if (year%4==0 and year%100!=0 )or (year%400==0):
        leap=True
    return leap

year = int(input())
print(is_leap(year))
```
